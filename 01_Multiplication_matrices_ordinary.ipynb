```python
import numpy as np
```

```python
def basic_matrix_product(A, B):
  # A , B = input matrices
    """ Basic_rule for matrix product."""
    # Rows and Columns for each matrix
    rows_A, cols_A = A.shape
    rows_B, cols_B = B.shape

    # Product_condition = Guard_rail
    assert cols_A == rows_B, "The #columns in A must equal to the #rows in B."

    # Set zero to all positions (both A & B)
    C = np.zeros((rows_A, cols_B))

    # Loops for Product
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                C[i, j] += A[i, k] * B[k, j]
                print (C[i,j],(i,j))

    return C
```

```python
# Example usage
A = np.array([[1, 2, 3, 2],
              [4, 5, 6, 2],
              [1, 2, 3, 2]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12],
              [1,1]])

# Simple method - two matrices product
C = basic_matrix_product(A, B)
print("Result of simple matrix product:\n", C)
```
