#### Este notebook versa Forca_Bruta com Memoizacao & Recursa para maxima_soma_array

```python
# import bibliotecas
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# parâmetros técnicos
N = np.random.randint(10,20) # tamanho de array - variacao aleatória
N
```

```python
# definicao de funcoes
def max_sum_memo(array):
  """ Esta funcao calcula com forca bruta a soma maxima, com recursao e memoizacao"""
  # array = input
  memo = {}
  def memos(i): # chama-se uma função dentro de outra
    if i == 0:
      return array [0] # para resolver os casos banais
    if i in memo:
      return memo[i]
    memo[i] = max(array[i], array[i] + memos(i-1)) # uso de recursao
    return memo[i] # uso de memoizacao
  return max(memos(i) for i in range(len(array)))
```

```python
# testes_simples
array1 = [10,-2,-4,3,2,6,-8,5]
array2 = [1,-1,2,-3]
max_sum_memo(array2)
```

```python
# dificuldade_computacional
array = np.random.randint(-100,100,N)
start = time.time_ns()
b = max_sum_memo(array)
end = time.time_ns()
duracao = end - start
duracao = round(duracao,2)
print(f'duracao em nanoseg =',duracao,'\narray_input =',array,'\n com max_soma =',b)
```

```python
# figuras
bucket_t = []
bucket_n = []
M = 20
for i in range(3,M):
  N = np.random.randint(3,100)
  array = np.random.randint(-20,100,N)
  start = time.time_ns()
  b = max_sum_memo(array)
  end = time.time_ns()
  duracao = end - start
  duracao = round(duracao,2)
  bucket_t.append(duracao)
  bucket_n.append(N)
plt.figure(figsize=(10,5))
plt.style.use('fivethirtyeight')
sns.scatterplot(x = bucket_n,y = bucket_t, color = 'blue', s =120)
plt.title('Esforço computacional')
plt.grid(True)
plt.xlabel('Tamanho do array')
plt.ylabel('Tempo de processamento em nanosegundos')
plt.show()
#
```
